```python
import json
```

```python
path = "../infra/contradictions3.json"

with open(path, "r") as f:
    data = json.load(f)
```

```python
contradictions = data["contradictions"]

c_references_none = 0
c_semantic_none = 0

c_references_total = 0
c_semantic_total = 0

unique = set()

for i in contradictions:
    evidence_a_bbox = i.get("evidence_a_bbox", None)
    evidence_b_bbox = i.get("evidence_b_bbox", None)
    edge_type = i.get("edge_type", "")

    unique.add(edge_type)

    if evidence_a_bbox is None or evidence_b_bbox is None:
        if edge_type == "reference":
            c_references_none += 1
        elif edge_type == "semantic_similarity":
            c_semantic_none += 1
    
    if edge_type == "reference":
        c_references_total += 1
    elif edge_type == "semantic_similarity":
        c_semantic_total += 1

len(contradictions), c_references_none, c_semantic_none, c_references_total - c_references_none, c_semantic_total - c_semantic_none, unique
```

```python
for c in range(len(data["edges"])):
  edge = data["edges"][c]
  nodes = data["nodes"]
  source_edges = edge["source"]
  target_edges = edge["target"]

  text_source = ""
  text_target = ""
  for i in nodes:
    id_ = i["id"]
    if id_ == source_edges:
      # print(id_)
      text_source = i["text"]
    if id_ == target_edges:
      # print(id_)
      text_target = i["text"]

  for j in contradictions:
    source = j["source"]
    target = j["target"]

    if source == source_edges and target == target_edges:
      evidence_a = j['evidence_a']
      evidence_b = j["evidence_b"]

      print(c, source_edges, target_edges)
      print("\n", text_source, "\n", text_target, "\n")
      print("\n", evidence_a, "\n", evidence_b, "\n")
      print("-----")
```
