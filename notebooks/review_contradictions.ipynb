```python
import json
```

```python
path = "../infra/contradictions2.json"

with open(path, "r") as f:
    data = json.load(f)
```

```python
contradictions = data["contradictions"]

c_references_none = 0
c_semantic_none = 0

c_references_total = 0
c_semantic_total = 0

for i in contradictions:
    evidence_a_bbox = i.get("evidence_a_bbox", None)
    evidence_b_bbox = i.get("evidence_b_bbox", None)
    edge_type = i.get("edge_type", "")

    if evidence_a_bbox is None or evidence_b_bbox is None:
        if edge_type == "reference":
            c_references_none += 1
        elif edge_type == "semantic_similarity":
            c_semantic_none += 1
    
    if edge_type == "reference":
        c_references_total += 1
    elif edge_type == "semantic_similarity":
        c_semantic_total += 1

len(contradictions), c_references_none, c_semantic_none, c_references_total - c_references_none, c_semantic_total - c_semantic_none
```
